```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Load the image
img = plt.imread(r"output.png")

rows, cols, _ = img.shape

plt.title("Original Image")
plt.imshow(img)
plt.show()

# Generate a random key
key = list(range(256))
random.shuffle(key)

def encrypt_image(img, key):
    img1 = np.zeros_like(img, dtype=int)
    for i in range(rows):
        for j in range(cols):
            for k in range(3):
                img1[i][j][k] = key[int(img[i][j][k] * 255)]
    return img1

img1 = encrypt_image(img, key)

plt.title("Encrypted Image")
plt.imshow(img1)
plt.show()

def decrypt_image(img, key):
    img1 = np.zeros_like(img, dtype=int)
    for i in range(rows):
        for j in range(cols):
            for k in range(3):
                img1[i][j][k] = key.index(int(img[i][j][k] * 255))
    return img1

img3 = decrypt_image(img1, key)

plt.title("Decrypted Image")
plt.imshow(img3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Load the image
img = plt.imread(r"output.png")

rows, cols, _ = img.shape

plt.title("Original Image")
plt.imshow(img)
plt.show()

# Generate a random key
key = list(range(256))
random.shuffle(key)

def encrypt_image(img, key):
    img1 = np.zeros_like(img, dtype=int)
    for i in range(rows):
        for j in range(cols):
            for k in range(3):
                img1[i][j][k] = key[int(img[i][j][k] * 255)]
    return img1 / 255.0  # Scale back to [0, 1]

img1 = encrypt_image(img, key)

plt.title("Encrypted Image")
plt.imshow(img1)
plt.show()

def decrypt_image(img, key):
    img1 = np.zeros_like(img, dtype=int)
    for i in range(rows):
        for j in range(cols):
            for k in range(3):
                img1[i][j][k] = key.index(int(img[i][j][k] * 255))
    return img1

img3 = decrypt_image(img1 * 255, key) / 255.0  # Scale back to [0, 1]

plt.title("Decrypted Image")
plt.imshow(img3)
plt.show()
```
